```python
# importing libraries 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import os
import sklearn
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.decomposition import PCA
```

```python
#importing the weather dataframe
path = r'C:\Users\karim\Documents\machine learning'
df_PCA = pd.read_csv(os.path.join(path, 'data', 'prepared Data', 'weather_PCA.csv'))
df_PCA
```

```python
#Dendrogram simgle method
PCA_single = linkage(df_PCA.loc[:,
['PCA1',
 'PCA2',
 'PCA3',
 'PCA4',
 'PCA5',
 'PCA6',
 'PCA7',
 'PCA8',
 'PCA9',
 'PCA10',
 'PCA11',
 'PCA12',
 ]],method="single")
plt.figure(figsize=(18,6))
dendrogram(PCA_single, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Dendrogram Single Method_PCA data",fontsize=18)
plt.show()
```

```python
#Dendrogram complete method
PCA_complete = linkage(df_PCA.loc[:,
['PCA1',
 'PCA2',
 'PCA3',
 'PCA4',
 'PCA5',
 'PCA6',
 'PCA7',
 'PCA8',
 'PCA9',
 'PCA10',
 'PCA11',
 'PCA12',
 ]],method="complete")
plt.figure(figsize=(18,6))
dendrogram(PCA_complete, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Dendrogram complete Method_PCA data",fontsize=18)
plt.show()
```

```python
#Dendrogram average method
PCA_average = linkage(df_PCA.loc[:,
['PCA1',
 'PCA2',
 'PCA3',
 'PCA4',
 'PCA5',
 'PCA6',
 'PCA7',
 'PCA8',
 'PCA9',
 'PCA10',
 'PCA11',
 'PCA12',
 ]],method="average")
plt.figure(figsize=(18,6))
dendrogram(PCA_average, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Dendrogram average Method_PCA data",fontsize=18)
plt.show()
```

```python
#Dendrogram ward method
PCA_ward = linkage(df_PCA.loc[:,
['PCA1',
 'PCA2',
 'PCA3',
 'PCA4',
 'PCA5',
 'PCA6',
 'PCA7',
 'PCA8',
 'PCA9',
 'PCA10',
 'PCA11',
 'PCA12',
 ]],method="ward")
plt.figure(figsize=(18,6))
dendrogram(PCA_ward, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Dendrogram ward Method_PCA data",fontsize=18)
plt.show()
```
